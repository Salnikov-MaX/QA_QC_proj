```python
class AlternativeData:
    def __init__(self, alternative_data: str):
        self.all_alternative_name = []
        for name_split in alternative_data.split('|/'):
            for name in name_split.split(','):
                self.all_alternative_name.append(name)

    def __str__(self):
        return f'({" || ".join(self.all_alternative_name)})'


class RequiredData:
    def __init__(self, required_data: str):
        self.required: [AlternativeData] = []
        for alternative_data in required_data.split('vs'):
            self.required.append(AlternativeData(alternative_data))

    def __str__(self):
        return "".join([i.all_alternative_name for i in self.required + [AlternativeData('')] ] + ['']) 
        

keys = []

for iii, data in enumerate(df2['Входные данные']):
    keys.append(RequiredData(data))
    print(data)
    print(RequiredData(data))
    print()
    
```
