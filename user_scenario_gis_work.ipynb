```python
import pandas as pd
import numpy as np

def color_cells(df, col_index_dict, color='red'):
    style = pd.DataFrame('', index=df.index, columns=df.columns)
    for col, rows in col_index_dict.items():
        for row in rows:
            style.loc[row, col] = f'background-color: {color}'
    return style

rows, cols = 20, 10
data = {f'Col_{i}': np.arange(1, rows + 1) for i in range(cols)}
df = pd.DataFrame(data)

color_dict = {
    'Col_1': [3, 10,],
    'Col_5': [1, 11, 14, 19],
    'Col_9': [0, 7, 19]
}

styled_df = df.style.apply(color_cells, axis=None, col_index_dict=color_dict, color='red')
display(styled_df)


```

```python
import pandas as pd
import numpy as np

def color_cells(df, col_index_color_dict):
    style = pd.DataFrame('', index=df.index, columns=df.columns)
    for col, rows in col_index_color_dict.items():
        for row, color in rows.items():
            style.loc[row, col] = f'background-color: {color}'
    return style


rows, cols = 20, 10
data = {f'Col_{i}': np.arange(1, rows + 1) for i in range(cols)}
df = pd.DataFrame(data)

color_dict = {
    'Col_1': {3: 'red', 10: 'blue', 17: 'green'},
    'Col_5': {1: 'yellow', 11: 'purple', 14: 'cyan'},
    'Col_9': {0: 'orange', 7: 'pink', 19: 'brown'}
}

styled_df = df.style.apply(color_cells, axis=None, col_index_color_dict=color_dict)
display(styled_df)
```

```python
import pandas as pd
import numpy as np

def color_specific_cells(df, col_index_color_dict):
    style = pd.DataFrame('', index=df.index, columns=df.columns)
    for col, rows in col_index_color_dict.items():
        for row, color in rows.items():
            style.loc[row, col] = f'background-color: {color}'
    return style

# Создаем синтетический датафрейм с рандомными значениями
rows, cols = 20, 10
data = {f'Col_{i}': np.random.randint(0, 100, size=rows) for i in range(cols)}
df = pd.DataFrame(data)

# Указываем, какие ячейки в каких столбцах раскрашивать и в какой цвет
color_dict = {
    'Col_1': {3: 'red', 10: 'blue', 17: 'green'},
    'Col_5': {1: 'yellow', 11: 'purple', 14: 'cyan'},
    'Col_9': {0: 'orange', 7: 'pink', 19: 'brown'}
}

styled_df = df.style.apply(color_specific_cells, axis=None, col_index_color_dict=color_dict)
display(styled_df)
```
